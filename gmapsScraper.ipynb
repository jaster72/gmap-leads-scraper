```python
from webdriver_manager.chrome import ChromeDriverManager

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

import time

from parsel import Selector

import re

from bs4 import BeautifulSoup
import requests

import csv

#initiate webdriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com")
#open specific page
bizType = input("What category do you want to search? ")
city = input("What city? ")
state = input("What state? ")
url = f'https://www.google.com/maps/search/{bizType}+in+{city}+{state}/'#@30.3831136,-97.7326164,13.65z'
driver.get(url)
#click into page and scroll to end of results
driver.find_element(By.XPATH,'//div[contains(@aria-label, "Results for")]/div/div[./a]').click()
actions = ActionChains(driver)
for _ in range(15):
    actions.send_keys(Keys.SPACE).perform()
    time.sleep(1)

#extract page contents
page_content = driver.page_source
soup = BeautifulSoup(page_content, "html.parser")

#structure data by company
bizList = soup.find_all('div',attrs={"jstcache": "80"})
leadsList = []
for biz in bizList:
    bizDetails = []
    bizDetails.append(biz.find('a')['aria-label'])
    try: 
        bizDetails.append(biz.find('a',attrs={"jstcache": "166"}).get('href'))
    except: 
        pass
    detailsList = biz.find_all('span',attrs={"jstcache": "140"})
    #print(len(detailsList))
    for item in detailsList:
        bizDetails.append(item.getText())
    print(bizDetails)
    leadsList.append(bizDetails)
print(leadsList)

#driver.quit()
```
